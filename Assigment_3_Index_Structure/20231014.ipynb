```python
from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups(subset="all")

print(len(data))
```

```python
query = "university"

for datum in data.data:
    if query in datum:
        print(datum)
    
```

```python
text_set = set()
for text in data.data:
    text_set.add(text)
    
print(len(text_set))
```

```python
query = data.data[0]

if query in text_set:
    print(query[:10])
else:
    print("no text")
```

```python
wordsets = []
for datum in data.data:
    wordsets.append(set(datum.split()))
print(len(wordsets))
```

```python
iindex = {}
i = 0
for wordset in wordsets:
    for word in wordset:
        if word in iindex:
            iindex[word].add(i)
        else:
            iindex[word] = {i}
    i += 1
print(len(iindex))
```

```python
query = "university"
print(iindex[query])
```

```python
q1 = "computer"
q2 = "science"
print(iindex[q1] & iindex[q2])
print(iindex[q1] | iindex[q2])
```
